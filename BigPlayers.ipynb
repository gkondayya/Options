```python
import pandas as pd
import datetime
```

```python
#url="https://raw.githubusercontent.com/gkondayya/Options/main/fao_participant_oi_25052022.csv"
url="https://www1.nseindia.com/content/nsccl/fao_participant_oi_26052022.csv"
df = pd.read_csv(url)
df.head()
```

```python
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header
df.head()
```

```python
df.dropna(axis=1)
```

```python
df['Future Stock Short\t']
```

```python

todays_date = datetime.datetime.now().date()
index = pd.date_range(todays_date, periods=1, freq='D')
columns = ["Date","FII_Fut","DII_Fut","PRO_Fut","FII_Opt","DII_Opt","PRO_Opt","NET_SMART_Fut","NET_SMART_Opt"]

#todays_date="2022-05-26"

# Creating a data frame
data = {'Date': [todays_date],
        'FII_Fut': [0],
        'DII_Fut': [0],
        'PRO_Fut': [0],
        'FII_Opt': [0],
        'DII_Opt': [0],
        'PRO_Opt': [0],
        'NET_SMART_Fut': [0],
        'NET_SMART_Opt': [0],
       }
  
df2 = pd.DataFrame(data, columns =columns,index=None)



df2
```

```python
df['Future Index Long'] = df['Future Index Long'].astype(int)
df['Future Index Short'] = df['Future Index Short'].astype(int)
df["Fut_OI"]=df['Future Index Long']-df['Future Index Short']

df2['FII_Fut'][0]=df['Fut_OI'][2]
df2['DII_Fut'][0]=df['Fut_OI'][3]
df2['PRO_Fut'][0]=df['Fut_OI'][4]
#df2['NET_SMART_Fut'][0] = df2['FII_Fut'][0]+df2['DII_Fut'][0]+df2['PRO_Fut'][0]

df2['NET_SMART_Fut'].iloc[0] = df2['FII_Fut'].iloc[0]+df2['DII_Fut'].iloc[0]+df2['PRO_Fut'].iloc[0]


df['Option Index Call Long'] = df['Option Index Call Long'].astype(int)
df['Option Index Call Short'] = df['Option Index Call Short'].astype(int)
df["Option_CALL_OI"]=df['Option Index Call Long']-df['Option Index Call Short']

df['Option Index Put Long'] = df['Option Index Put Long'].astype(int)
df['Option Index Put Short'] = df['Option Index Put Short'].astype(int)
df["Option_Put_OI"]=df['Option Index Put Long']-df['Option Index Put Short']

df['Option_OI']=df["Option_CALL_OI"]-df["Option_Put_OI"]

df2['FII_Opt'][0]=df['Option_OI'][2]
df2['DII_Opt'][0]=df['Option_OI'][3]
df2['PRO_Opt'][0]=df['Option_OI'][4]
df2['NET_SMART_Opt'].iloc[0]= df2['FII_Opt'].iloc[0]+df2['DII_Opt'].iloc[0]+df2['PRO_Opt'].iloc[0]

df.head()
```
